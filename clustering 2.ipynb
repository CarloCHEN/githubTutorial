```python
X_train = subjects2[subjects2['subject_number']!=10].iloc[:,1:29]
X_test = subjects2[subjects2['subject_number']==10].iloc[:,1:29]

y_train = subjects2[subjects2['subject_number']!=10].iloc[:,29] # first attribute
y_test = subjects2[subjects2['subject_number']==10].iloc[:,29]

scaler = StandardScaler()
scaler.fit(X_train)
X_scaled_train = scaler.transform(X_train)
X_scaled_test = scaler.transform(X_test)
```

```python
DT_model = DecisionTreeClassifier()
DT_model.fit(X_train,y_train)
DT_model.score(X_test,y_test)
```

```python
KNN_model = KNeighborsClassifier()
KNN_model.fit(X_scaled_train,y_train)
KNN_model.score(X_scaled_test,y_test)
```
